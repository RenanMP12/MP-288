```python
def coef_(y_):
    x = np.array([0, 1, 2, 3, 4, 5, 6])
    y = np.array([0.5, 0.5, y_[0], y_[1], y_[2], y_[3], 1.0])
    s = np.zeros(2)
    
    #     initiating matrix a & b
    a = np.zeros((24,24))
    b = np.zeros(24)
    
    #     C0 continuity conditions
    a[[0],[0]] = 1
    a[[0],[1]] = x[0]
    a[[0],[2]] = x[0]**2
    a[[0],[3]] = x[0]**3
    
    a[[1],[0]] = 1
    a[[1],[1]] = x[1]
    a[[1],[2]] = x[1]**2
    a[[1],[3]] = x[1]**3
    
    a[[2],[4]] = 1
    a[[2],[5]] = x[1]
    a[[2],[6]] = x[1]**2
    a[[2],[7]] = x[1]**3
    
    a[[3],[4]] = 1
    a[[3],[5]] = x[2]
    a[[3],[6]] = x[2]**2
    a[[3],[7]] = x[2]**3
    
    a[[4],[8]]  = 1
    a[[4],[9]]  = x[2]
    a[[4],[10]] = x[2]**2
    a[[4],[11]] = x[2]**3
    
    a[[5],[8]]  = 1
    a[[5],[9]]  = x[3]
    a[[5],[10]] = x[3]**2
    a[[5],[11]] = x[3]**3
    
    a[[6],[12]] = 1
    a[[6],[13]] = x[3]
    a[[6],[14]] = x[3]**2
    a[[6],[15]] = x[3]**3
    
    a[[7],[12]] = 1
    a[[7],[13]] = x[4]
    a[[7],[14]] = x[4]**2
    a[[7],[15]] = x[4]**3
    
    a[[8],[16]] = 1
    a[[8],[17]] = x[4]
    a[[8],[18]] = x[4]**2
    a[[8],[19]] = x[4]**3    
    
    a[[9],[16]] = 1
    a[[9],[17]] = x[5]
    a[[9],[18]] = x[5]**2
    a[[9],[19]] = x[5]**3  
    
    a[[10],[20]] = 1
    a[[10],[21]] = x[5]
    a[[10],[22]] = x[5]**2
    a[[10],[23]] = x[5]**3  
    
    a[[11],[20]] = 1
    a[[11],[21]] = x[6]
    a[[11],[22]] = x[6]**2
    a[[11],[23]] = x[6]**3
    
    #     C1 continuity conditions
    a[[12], [0]] = 0
    a[[12], [1]] = 1
    a[[12], [2]] = 2*x[1]
    a[[12], [3]] = 3*x[1]**2
    
    a[[12], [4]] =  0
    a[[12], [5]] = -1
    a[[12], [6]] = -2*x[1]
    a[[12], [7]] = -3*x[1]**2    
    
    a[[13], [4]] = 0
    a[[13], [5]] = 1
    a[[13], [6]] = 2*x[2]
    a[[13], [7]] = 3*x[2]**2
    
    a[[13], [8]] =  0
    a[[13], [9]] = -1
    a[[13],[10]] = -2*x[2]
    a[[13],[11]] = -3*x[2]**2
    
    a[[14], [8]] = 0
    a[[14], [9]] = 1
    a[[14],[10]] = 2*x[3]
    a[[14],[11]] = 3*x[3]**2
    
    a[[14],[12]] =  0
    a[[14],[13]] = -1
    a[[14],[14]] = -2*x[3]
    a[[14],[15]] = -3*x[3]**2
    
    a[[15],[12]] = 0
    a[[15],[13]] = 1
    a[[15],[14]] = 2*x[4]
    a[[15],[15]] = 3*x[4]**2
    
    a[[15],[16]] =  0
    a[[15],[17]] = -1
    a[[15],[18]] = -2*x[4]
    a[[15],[19]] = -3*x[4]**2
    
    a[[16],[16]] = 0
    a[[16],[17]] = 1
    a[[16],[18]] = 2*x[5]
    a[[16],[19]] = 3*x[5]**2
    
    a[[16],[20]] =  0
    a[[16],[21]] = -1
    a[[16],[22]] = -2*x[5]
    a[[16],[23]] = -3*x[5]**2
    
    #     C2 continuity conditions
    a[[17], [0]] = 0
    a[[17], [1]] = 0
    a[[17], [2]] = 2
    a[[17], [3]] = 6*x[1]
    
    a[[17], [4]] =  0
    a[[17], [5]] =  0
    a[[17], [6]] = -2
    a[[17], [7]] = -6*x[1]
    
    a[[18], [4]] = 0
    a[[18], [5]] = 0
    a[[18], [6]] = 2
    a[[18], [7]] = 6*x[2]
    
    a[[18],  [8]] =  0
    a[[18],  [9]] =  0
    a[[18], [10]] = -2
    a[[18], [11]] = -6*x[2]
    
    a[[19],  [8]] = 0
    a[[19],  [9]] = 0
    a[[19], [10]] = 2
    a[[19], [11]] = 6*x[3]
    
    a[[19], [12]] =  0
    a[[19], [13]] =  0
    a[[19], [14]] = -2
    a[[19], [15]] = -6*x[3]
    
    a[[20], [12]] = 0
    a[[20], [13]] = 0
    a[[20], [14]] = 2
    a[[20], [15]] = 6*x[4]
    
    a[[20], [16]] =  0
    a[[20], [17]] =  0
    a[[20], [18]] = -2
    a[[20], [19]] = -6*x[4]
    
    a[[21], [16]] = 0
    a[[21], [17]] = 0
    a[[21], [18]] = 2
    a[[21], [19]] = 6*x[5]
    
    a[[21], [20]] =  0
    a[[21], [21]] =  0
    a[[21], [22]] = -2
    a[[21], [23]] = -6*x[5]
    
    #     initial and final slope conditions
    a[[22],[0]] = 0
    a[[22],[1]] = 1
    a[[22],[2]] = 2*x[0]
    a[[22],[3]] = 3*x[0]**2
    
    a[[23],[20]] = 0
    a[[23],[21]] = 1
    a[[23],[22]] = 2*x[6]
    a[[23],[23]] = 3*x[6]**2
    
    #     define vector b
    b[[0]] = y[0]
    b[[1]] = y[1]
    b[[2]] = y[1]
    b[[3]] = y[2]
    b[[4]] = y[2]
    b[[5]] = y[3]    
    b[[6]] = y[3]    
    b[[7]] = y[4]
    b[[8]] = y[4]
    b[[9]] = y[5]
    b[[10]] = y[5]
    b[[11]] = y[6]
    b[[22]] = s[0]
    b[[23]] = s[1]    
    
    #     solve linear equation system
    c = np.linalg.solve(a, b)
    
    return c
   
def area(c):
    x = np.array([0, 1, 2, 3, 4, 5, 6])
    #     area under spline
    
    Ar_ = np.zeros(6)
    
    for i in range(6):
        Ar_[i] = ( c[4*i]*x[i+1] + 1/2*c[4*i + 1]*x[i+1]**2 + 1/3*c[4*i + 2]*x[i+1]**3 + 1/4*c[4*i + 3]*x[i+1]**4)- (c[4*i]*x[i] + 1/2*c[4*i + 1]*x[i]**2 + 1/3*c[4*i + 2]*x[i]**3 + 1/4*c[4*i + 3]*x[i]**4)
        
    Ar = np.sum(Ar_)
    
    return Ar

def revolution_area(c):
    x = np.array([0, 1, 2, 3, 4, 5, 6])
    #  revolution surface area
    
    Area_ = np.zeros(6)
    
    #  number of sampling points at gauss-legendre quadrature
    n = 10 
    xi, wi = np.polynomial.legendre.leggauss(n)
    
    #  does for each i-th interval
    for i in range(6):
        F   = np.zeros(n)
        Fw  = np.zeros(n)
        eps = np.zeros(n)
        
        for j in range(n):
            #  finds epsilon sampling points for the interval
            eps[j] = (x[i+1] + x[i])/2 + (x[i+1] - x[i])/2*xi[j]
            
            #  evaluates f(eps)*sqrt[1+(f(eps)')^2]
            F[j] = (c[4*i] + c[4*i + 1]*eps[j] + c[4*i + 2]*eps[j]**2 + c[4*i + 3]*eps[j]**3)*np.sqrt( 1 + (c[4*i+1]+2*c[4*i+2]*eps[j]+3*c[4*i+3]*eps[j]**2)**2)
            
            #  evaluates F(j) times integration weights wi
            Fw[j] = (x[i+1] - x[i])/2*F[j]*wi[j]
        
        Area_[i] = 2*np.pi*np.sum(Fw)
        
    Area = np.sum(Area_)
    
    return Area

def revolution_volume(c):
    x = np.array([0, 1, 2, 3, 4, 5, 6])
    # the revolution volume of the splines
    
    volume_ = np.zeros(6)
    
    #  number of sampling points at gauss-legendre quadrature
    n = 4
    xi, wi = np.polynomial.legendre.leggauss(n)
    
    #  does for each i-th interval
    for i in range(6):
        F   = np.zeros(n)
        Fw  = np.zeros(n)
        eps = np.zeros(n)
        
        for j in range(n):
            #  finds epsilon sampling points for the interval
            eps[j] = (x[i+1] + x[i])/2 + (x[i+1] - x[i])/2*xi[j]
            
            #  evaluates [f(eps)]^2
            F[j] = (c[4*i] + c[4*i + 1]*eps[j] + c[4*i + 2]*eps[j]**2 + c[4*i + 3]*eps[j]**3)**2
            
            #  evaluates F(j) times integration weights wi
            Fw[j] = (x[i+1] - x[i])/2*F[j]*wi[j]
        
        volume_[i] = np.pi*np.sum(Fw)    
        
    volume = np.sum(volume_)
    
    return volume

def length(c):
    x = np.array([0, 1, 2, 3, 4, 5, 6])
    #  calculates length integrals numerically
    length_ = np.zeros(6)
        
    #  number of sampling points at gauss-legendre quadrature
    n = 10
    xi, wi = np.polynomial.legendre.leggauss(n)
    
    #  does for each i-th interval
    for i in range(6):
        F   = np.zeros(n)
        Fw  = np.zeros(n)
        eps = np.zeros(n)
        
        for j in range(n):
            #  finds epsilon sampling points for the interval
            eps[j] = (x[i+1] + x[i])/2 + (x[i+1] - x[i])/2*xi[j]
            
            #  evaluates sqrt[1+(f(eps)')^2]
            F[j] = np.sqrt(1+(c[4*i + 1]+2*c[4*i + 2]*eps[j]+3*c[4*i + 3]*eps[j]**2)**2)
            
            #  evaluates F(j) times integration weights wi
            Fw[j] = (x[i+1] - x[i])/2*F[j]*wi[j]
        
        length_[i] = np.sum(Fw)    
        
    length = np.sum(length_)
    
    return length

def plot_spl(c):
    x = np.array([0, 1, 2, 3, 4, 5, 6])
    #  plot the splines
    xx1 = np.linspace(x[0], x[1], num = 10)
    xx2 = np.linspace(x[1], x[2], num = 10)
    xx3 = np.linspace(x[2], x[3], num = 10)
    xx4 = np.linspace(x[3], x[4], num = 10)
    xx5 = np.linspace(x[4], x[5], num = 10)
    xx6 = np.linspace(x[5], x[6], num = 10)
    
    yy1 = c[0]+c[1]*xx1+c[2]*xx1**2+c[3]*xx1**3
    yy2 = c[4]+c[5]*xx2+c[6]*xx2**2+c[7]*xx2**3
    yy3 = c[8]+c[9]*xx3+c[10]*xx3**2+c[11]*xx3**3
    yy4 = c[12]+c[13]*xx4+c[14]*xx4**2+c[15]*xx4**3
    yy5 = c[16]+c[17]*xx5+c[18]*xx5**2+c[19]*xx5**3
    yy6 = c[20]+c[21]*xx6+c[22]*xx6**2+c[23]*xx6**3
    
    plt.plot(xx1,yy1,xx2,yy2,xx3,yy3,xx4,yy4,xx5,yy5,xx6,yy6,0,0,1,1)
    plt.xlim([0,6])
    plt.ylim([0,2])
    plt.grid()
    
def gradient(f, x, h = 1e-4):                             # gradient function
    df = np.zeros(len(x))
    for i in range(0, len(x)):
        xp    = np.array(x, dtype=float)
        xp[i] = x[i] + h
        df[i] = (f(xp) - f(x))/h
    
    return df

def golden_method(f, delta, x0, epsilon):                    # Golden section method
    gsm = (math.sqrt(5) + 1)/2
    
    # Phase 01
    
    f_0 = f(x0)
    counter = 0
    cond = True
    gradf = -gradient(f, x0)
    alfa_m = np.zeros(3)
    alfa = 0
    
    while cond:       
        alfa = alfa + delta*gsm**counter
        x = np.zeros(len(x0))
        for i in range(0, len(x0)):
            x[i] = x0[i] + + alfa*gradf[i]
        f_1 = f(x)
        
        if f_1 < f_0:
            f_0 = f_1
            counter = counter + 1
            alfa_m[0] = alfa_m[1]
            alfa_m[1] = alfa_m[2]
            alfa_m[2] = alfa
        else:
            counter = 0    
            cond = False
            alfa_m[0] = alfa_m[1]
            alfa_m[1] = alfa_m[2]
            alfa_m[2] = alfa  
            
     # Phase 02
    
    alfa_range = np.array([alfa_m[0], alfa_m[2]])
    cond = True
    
    while cond:
        counter = counter + 1
        
        I = alfa_range[1] - alfa_range[0]
        
        alfa_a = alfa_range[0] + (1 - 1/gsm)*I
        
        x_a = np.zeros(len(x0))
        for i in range(0, len(x0)):
            x_a[i] = x0[i] + + alfa_a*gradf[i]
        
        f_a = f(x_a)
        
        alfa_b = alfa_range[0] + 1/gsm*I
        
        x_b = np.zeros(len(x0))
        for i in range(0, len(x0)):
            x_b[i] = x0[i] + + alfa_b*gradf[i]
        
        f_b = f(x_b)
    
        
        if f_a > f_b and I > epsilon:
            alfa_range[0] = alfa_a
            
        elif f_a < f_b and I > epsilon:
            alfa_range[1] = alfa_b
            
        elif I < epsilon:
            cond = False
            alfa = (alfa_a + alfa_b)/2
    
    return alfa
```
