```python
from sympy import *
init_printing(use_unicode=True)
x, y = symbols('x y')
partial = Matrix([[x],[y]])                        # gradiente da função objetivo
```

```python
f = 4*(sqrt( x**2 + (10 - y)**2) - 10)**2 + 1/2*(sqrt( x**2 + (10 + y)**2) - 10)**2 - 5*(x + y)
f
```

```python
gradf = diff(f,partial)
N(gradf.subs({x: 8.48, y: 4.42}))
```

```python
g1 = x + y**2/6 - 12
g1.subs({x: 8.48, y: 4.42})
```

```python
gradg1 = diff(g1,partial)
N(gradg1.subs({x: 8.48, y: 4.42}))
```

```python
g2 = 5*x - y - 38
g2.subs({x: 8.48, y: 4.42})
```

```python
gradg2 = diff(g2,partial)
N(gradg2.subs({x: 8.48, y: 4.42}))
```

```python
N(gradf.subs({x: 8.48, y: 4.42})) + 0.05*N(gradg2.subs({x: 8.48, y: 4.42}))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# variaveis para minimização
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
d1, d2 = np.meshgrid(x, y)
```

```python
# Lagrangiana
L  = -0.33*d1+0.09*d2 + 0.5*(d1*(d1*6.17 - d2*3.36)+d2*(-d1*3.36+d2*3.35))
g1 = -0.27+d1+1.47*d2
g2 = -0.02 + 5*d1-d2
```

```python
fig = plt.figure()
plt.grid(b=True, which='major', color='black', linestyle='--');
contours = plt.contour(d1, d2, L, 10)                            # plot função objetivo
plt.clabel(contours, inline=True, fontsize=20)

plt.contour(d1, d2, g1, 0, colors = 'red',  linestyles='dashed')     
plt.contour(d1, d2, g2, 0, colors = 'blue', linestyles='dashed')  
plt.plot(0,0,color='green',marker='o',markersize=12)
fig.savefig('lista_07_iteracao_05.png',dpi=300)                              # salvar figura como png (dpi = número de pixels)
plt.xticks(Fontsize = 25)
plt.yticks(Fontsize = 25)
fig.set_size_inches(20, 12)
plt.show()
```
